```
import pandas as pd
!pip install Biopython
from Bio import SeqIO
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
import matplotlib.pyplot as plt
import random
```

```
ppi_data = pd.read_csv("/content/ppi_multipathogen_wCoV2.txt",sep='\t')
ppi_data.dataset.unique()
ppi_data[ppi_data.dataset=='SARS-CoV2-HEK293T'].Bait.unique()
ppi_data = ppi_data.reset_index(drop=True)
```

```
sars_cov2 = pd.read_csv("/content/sarscov2_seq (1).txt",sep='\t')
sars_cov2.drop(sars_cov2.index[:1,], inplace=True)
sars_cov2.drop(sars_cov2.index[2:], inplace=True)
sars_cov2.columns = ['Bait']
sars_cov2['Bait_Sequence'] = sars_cov2.loc[2, 'Bait']
sars_cov2.drop(sars_cov2.index[1], inplace=True)
sars_cov2.reset_index(drop=True, inplace=True)
sars_cov2.loc[0, 'Bait'] = 'SARS-CoV2 M'
sars_cov2
```

```
human = {'Prey': [], 'Prey_Sequence': []}
for record in SeqIO.parse("/content/uniprot-proteome%3AUP000005640.fasta", "fasta"):
    human['Prey'].append(record.id.split("|")[1])
    human['Prey_Sequence'].append(str(record.seq))
human = pd.DataFrame(human, columns=['Prey', 'Prey_Sequence'])
human
```

```
sars_cov2_ppi = ppi_data[(ppi_data['pathogen']=='SARS-CoV2')]

sars_cov2_ppi.drop(sars_cov2_ppi.columns[5:], axis = 1, inplace = True)
sars_cov2_ppi.drop(sars_cov2_ppi.columns[:3], axis = 1, inplace = True)   

sars_cov2_ppi['Bait_Sequence']=sars_cov2_ppi['Bait'].map(sars_cov2.set_index('Bait')['Bait_Sequence'])
sars_cov2_ppi['Prey_Sequence']=sars_cov2_ppi['Prey'].map(human.set_index('Prey')['Prey_Sequence'])

sars_cov2_ppi = sars_cov2_ppi.dropna()
sars_cov2_ppi.reset_index(inplace=True, drop=True)

sars_cov2_ppi
```

```
column_names = ["Bait", "Bait_Sequence", "Prey", "Prey_Sequence", "Scores"]

sars_cov2_df = pd.DataFrame(columns = column_names)

for i in range(len(human)):
  X = sars_cov2_ppi["Bait_Sequence"].unique()
  Y = human["Prey_Sequence"][i]

  alignments = pairwise2.align.globalxx(X, Y)

  sars_cov2_df.loc[len(sars_cov2_df)] = ['M', X, human['Prey'][i], Y, alignments[0][2]]

```

```
SARS_CoV2_Text = sars_cov2_df.to_csv('SARS_CoV2_Text.txt', sep='\t', index=False)
```
